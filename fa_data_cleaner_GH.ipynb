```python
import pandas as pd
import sys

# Define constants for input and output files
INPUT_FILE = 'FA.xlsx'
OUTPUT_FILE = 'FA_Updated.xlsx'

def run_automation():
    """
    Main function to load, clean, filter, and preprocess payment reports.
    """
    try:
        # 1. Load the raw Excel report into a pandas DataFrame
        df = pd.read_excel(INPUT_FILE)
        print(f"✅ Success: '{INPUT_FILE}' read successfully!")

        # 2. Convert date strings to datetime objects and sort chronologically
        df['Fee Date'] = pd.to_datetime(df['Fee Date'])
        df = df.sort_values(by='Fee Date', ascending=True)
        print("Data sorted.")

        # 3. Filter out unnecessary rows (e.g., specific clients not needed for this workflow)
        df = df[df['Client Code'] != 'XXXX']  # Example: Remove rows for client code 'XXXX'
        print("Removed unnecessary rows.")

        # 4. Filter for records that require active processing (Status: Pending)
        df = df[df['Fee Status'] == 'Pending']
        print("Status filtered.")

        # 5. Keep only records where 'Actioned By' is blank/NaN (indicating no previous attempts)
        df = df[df['Actioned By'].isna()]
        print("Removed rows with no attempts.")

        # 6. Select and retain only the essential columns required for the final task
        columns_to_keep = [
            "COLUMN1", "COLUMN2", "COLUMN3"
        ]
        df = df[columns_to_keep]
        print("Selected final columns.")

        # 7. Export the clean data to a new Excel file (excluding the pandas index column)
        df.to_excel(OUTPUT_FILE, index=False)
        print(f"Done! Final file saved as: {OUTPUT_FILE}")
        print(f"Rows left to work on: {len(df)}")

        # Print script metadata and ownership
        print("\n" + "="*40)
        print("© 2026 QB. All Rights Reserved.")
        print("This automation script was custom-built by Mohammad Sas.")
        print("="*40)

    except FileNotFoundError:
        # Handle case where the raw report is missing or misplaced
        print(f"❌ Error: The file '{INPUT_FILE}' was not found.")
    except PermissionError:
        # Handle case where the output file is currently open in Excel
        print(f"❌ Error: Please close '{OUTPUT_FILE}' in Excel and try again.")
    except KeyError as e:
        # Handle mismatch in column names (e.g., if raw report layout changes)
        print(f"❌ Column Error: Check spelling of {e}")
    except Exception as e:
        # Fallback for any other unexpected runtime exceptions
        print(f"❌ An unexpected error occurred: {e}")

if __name__ == "__main__":
    run_automation()
```
